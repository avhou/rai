```python
from wordcloud import WordCloud
from typing import *
import matplotlib.pyplot as plt
from utils import read_dataset, tokenize

def show_wordcloud(data: str, title: str, output_file: str):
    wordcloud = WordCloud(
        background_color='white',
        max_words=100,
        max_font_size=40,
        scale=3,
        random_state=1
    ).generate(data)

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title:
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    wordcloud.to_file(output_file)
    plt.show()

    
datasources = read_dataset("dataset.csv")

show_wordcloud(" ".join([source.text for source in datasources if source.category == "scientific-before"]), "scientific community before ChatGPT", "figures/scientific-before.png")
show_wordcloud(" ".join(tokenize(" ".join([source.text for source in datasources if source.category == "scientific-before"]))), "scientific community before ChatGPT", "figures/scientific-before-tokenized.png")
show_wordcloud(" ".join([source.text for source in datasources if source.category == "scientific-after"]), "scientific community after ChatGPT", "figures/scientific-after.png")
show_wordcloud(" ".join(tokenize(" ".join([source.text for source in datasources if source.category == "scientific-after"]))), "scientific community after ChatGPT", "figures/scientific-after-tokenized.png")
show_wordcloud(" ".join([source.text for source in datasources if source.category == "non-scientific-before"]), "non scientific community before ChatGPT", "figures/non-scientific-before.png")
show_wordcloud(" ".join(tokenize(" ".join([source.text for source in datasources if source.category == "non-scientific-before"]))), "non scientific community before ChatGPT", "figures/non-scientific-before-tokenized.png")
show_wordcloud(" ".join([source.text for source in datasources if source.category == "non-scientific-after"]), "non scientific community after ChatGPT", "figures/non-scientific-after.png")
show_wordcloud(" ".join(tokenize(" ".join([source.text for source in datasources if source.category == "non-scientific-after"]))), "non scientific community after ChatGPT", "figures/non-scientific-after-tokenized.png")

print("Done.")
```
